We will use the "House Sales in King County, USA" dataset from Kaggle.
<br>Kaggle dataset link:
<br>https://www.kaggle.com/datasets/harlfoxem/housesalesprediction

```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
import os
```

```python
# The os.getcwd() function is used to retrieve the current working directory of your script file or environment.
CURRENT_DIR = os.getcwd()
PARENT_DIR = os.path.dirname(CURRENT_DIR)
GET_CSV_FILE = os.path.join(PARENT_DIR, "03.Pandas/kc_house_data _rev.csv")

print(CURRENT_DIR)
print(PARENT_DIR)
print(GET_CSV_FILE)
```

```python
df_houses = pd.read_csv(GET_CSV_FILE)
```

```python
sort_df_by_year = df_houses.sort_values(by=["yr_built"])

with pd.option_context('display.max_columns', None):
    display(sort_df_by_year.head())
```

#### **Adding Labels to a Plot:**

```python
years_built = sort_df_by_year["yr_built"]
prices = sort_df_by_year["price"]

plt.plot(years_built, prices)

plt.xlabel("Year Built")
plt.ylabel("Price")

plt.show()
```

#### **Adding Title to a Plot:**

```python
years_built = sort_df_by_year["yr_built"]
prices = sort_df_by_year["price"]

plt.plot(years_built, prices)

plt.title("Houses Prices")
plt.xlabel("Year Built")
plt.ylabel("Price")

plt.show()
```

#### **Setting Font Properties for the Title and Labels:**
For more information about text properties and layout, check this link:
<br>https://matplotlib.org/stable/users/explain/text/text_props.html#text-props

```python
years_built = sort_df_by_year["yr_built"]
prices = sort_df_by_year["price"]

plt.plot(years_built, prices)

title_font_property = {"family": "serif", "style": "italic", "color": "green", "size": 25, "weight": "bold"}
labels_font_property = {"family": "sans-serif", "style": "italic", "color": "0.5", "size": 15, "weight": "bold"}

plt.title("Houses Prices", fontdict=title_font_property)
plt.xlabel("Year Built", fontdict=labels_font_property)
plt.ylabel("Price", fontdict=labels_font_property)

plt.show()
```

```python
years_built = sort_df_by_year["yr_built"]
prices = sort_df_by_year["price"]

plt.plot(years_built, prices)

title_font_property = {"family": "serif", "style": "italic", "color": "green", "size": 25, "weight": "bold",}
labels_font_property = {"family": "sans-serif", "style": "italic", "color": "0.5", "size": 15, "weight": "bold"}

plt.title("Houses Prices", fontdict=title_font_property, loc="left")
plt.xlabel("Year Built", fontdict=labels_font_property)
plt.ylabel("Price", fontdict=labels_font_property)

plt.show()
```

#### **Adding Grid Lines to a Plot:**

```python
years_built = sort_df_by_year["yr_built"]
prices = sort_df_by_year["price"]

plt.plot(years_built, prices)

title_font_property = {"family": "serif", "style": "italic", "color": "green", "size": 25, "weight": "bold",}
labels_font_property = {"family": "sans-serif", "style": "italic", "color": "0.5", "size": 15, "weight": "bold"}

plt.title("Houses Prices", fontdict=title_font_property, loc="right")
plt.xlabel("Year Built", fontdict=labels_font_property)
plt.ylabel("Price", fontdict=labels_font_property)

# plt.grid()
# plt.grid(axis="x")
plt.grid(axis="y")
plt.show()
```
