##### **What are quantiles?**
Quantiles are values that divide sorted data or a probability distribution into equal parts.
<br>Quartiles are specific percentiles that divide a dataset into four equal parts.
<br>A percentile indicates that a given percentage of data points in a set fall at or below that specific value
<br> To get the quantiles of a specific data set, we use the "quantile" method
<br> The "quantile" method takes values between 0 and 1.

```python
# The first quartile, Q1, is the 25th percentile, which means 25% of the data falls below Q1.
price_Q1 = df_houses["price"].quantile(0.25)
print(price_Q1)
```

```python
# The second quartile, Q2, is the median, which is the 50th percentile, meaning 50% of the data falls below Q2.
# The default value in the "quantile" method is "0.5":
price_Q2 = df_houses["price"].quantile()
print(price_Q2)
```

```python
# The third quartile, Q3, is the 75th percentile, which means 75% of the data falls below Q3.
price_Q3 = df_houses["price"].quantile(0.75)
print(price_Q3)
```

#### **Aggregate method:**
The "agg" method in pandas is used to apply one or more aggregation functions or methods across an entire DataFrame or specific columns.

```python
df_houses[["price", "sqft_living"]].agg(["max", "mean", "quantile"])
```

User-defined functions can be use:

```python
def quantiles_cal(column, *args):
    quntile_values = tuple(column.quantile(q) for q in args)
    return quntile_values

# Setting the dataframe index
quantile_index = pd.Series(["Q1", "Q2", "Q3"])

# Apply the user-defined function using agg.
# The 0 argument in the "agg" method is the default axis; it should be added if arbitrary arguments are used. 
quantile_result = df_houses[["price", "sqft_living"]].agg(quantiles_cal, 0, 0.25, 0.5, 0.75).set_index(quantile_index)
print(quantile_result)
```

#### **Dropping Duplicates:**
"drop_duplicates" method returns a DataFrame with duplicate rows removed.

```python
# Let's take an example
new_df = df_houses.loc[100:150, "price": "yr_built"]
```

```python
# Let's say we want to remove the duplicate years:
drop_br = new_df.drop_duplicates(subset="bedrooms")

drop_br
```

```python
# Using multiple subsets will remove rows that have duplicates from both columns:
drop_br_floors = new_df.drop_duplicates(subset=["bedrooms", "floors"])

drop_br_floors
```

#### **Grouping Data:**
A groupby operation involves: splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and apply operations on these groups.

```python
# Group by the "yr_built" and calculate the mean for all the data at a specific year
sum_year = df_houses.groupby("yr_built").sum(numeric_only=True)

sum_year.head(10)
```

```python
avg_year_prices = df_houses.groupby("yr_built")["price"].mean(numeric_only=True)

avg_year_prices.head(10)
```

```python
year_prices_sqli = df_houses.groupby("yr_built")[["price", "sqft_living"]].median(numeric_only=True)

year_prices_sqli.head(10)
```

```python
# Applying multiple functions using the groupby "method":
# diif_per = df_houses.groupby("yr_built")["price"].agg(["max", "mean", "min"])
diif_per = df_houses.groupby("yr_built")[["price", "sqft_living"]].agg(["max", "mean", "min"])

diif_per.head(10)
```

```python
# Grouping by multiple columns:
multiple_columns = df_houses.groupby(["yr_built", "grade"])[["price", "sqft_living"]].mean(numeric_only=True)

multiple_columns.head(20)
```
