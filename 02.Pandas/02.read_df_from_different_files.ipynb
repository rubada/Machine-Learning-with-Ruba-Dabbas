You can use Pandas to read and write to different data files, and you can also use Pandas to connect to the database and read or write to it.

To use Pandas to read from different files there are functions in Pandas that are used to read from these files:
<br> 1. To read from CSV files, use the "*pd.read_csv*".
<br> 2. To read from Excel files, use the "*pd.read_excel*".
<br> 2. To read from JSON files, use the "*pd.read_json*".

To use Pandas to write to different files there are functions in Pandas that are used to write to these files:
<br> 1. To write to CSV files, use the "*pd.DataFrame.to_csv*".
<br> 2. To write to Excel files, use the "*pd.DataFrame.to_excel*".
<br> 2. To write to JSON files, use the "*pd.DataFrame.to_json*".

For more information about all the files that Pandas can read and write to, check out this link:
<br>https://pandas.pydata.org/docs/reference/io.html

Let's take an example and check out how we can read a file, we will work with "CSV" files:

We will use the "House Sales in King County, USA" dataset from Kaggle.
<br>Kaggle dataset link:
<br>https://www.kaggle.com/datasets/harlfoxem/housesalesprediction

The "read_csv" function has many parameters, we will not discuss them all, but you can check them out in the below link:
<br>https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html#pandas.read_csv

We will use the following:
<br>***filepath_or_buffer:***" It is the name of the CSV file, you can add the absolute or relative path or the CSV file name if the CSV file and the script file are in the same folder.
<br>***sep:*** It is the separator between the columns in a CSV file, the default is ",", if other characters separate the columns you should use the "sep" parameter and add this character as a string.

```python
import pandas as pd
```

### **Reading a CSV File:**

```python
df = pd.read_csv("kc_house_data.csv")
```

```python
# The "head" function returns the first n rows in the Dataframe or Series.
# It is useful for quickly checking out the data.
# The default number of rows is 5:
df.head()
```

```python
df.head(10)
```

```python
# The "tail" function returns the last n rows in the Dataframe or Series.
# It is useful for quickly checking the data after sorting, adding, removing rows, etc.
# The default number of rows is 5:

df.tail()
```

```python
df.tail(7)
```

As shown above, when we print the DataFrame head and the tail it didn't show all the columns, to show all the DataFrame columns we should use the following function:
<br>**pandas.set_option(pat, value)**
<br>It is used to set the value of the specified option, check below link for more information:
<br>https://pandas.pydata.org/docs/reference/api/pandas.set_option.html#pandas.set_option

It has two parameters:
<br>***pat:*** It is a string and should be equal to "display.max_columns".
<br>***value:*** It is a number if a number is specified to it, it will return a Dataframe with column numbers equal to "value".
<br>If the "value is "None" it will return all the Dataframe columns.

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

To temporarily show all columns without changing the global setting:

```python
with pd.option_context('display.max_columns', None):
    print(df.head())
```

```python
df.head()
```

When working with big data, using the "print" function to display the data displayed will be messy, as shown above.
<br> Instead of using the "print" function, we can use the "display" function from the "IPython.display" module.

```python
from IPython.display import display
```

```python
with pd.option_context('display.max_columns', None):
    display(df.head())
```
