```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
CURRENT_DIR = os.getcwd()
PARENT_DIR = os.path.dirname(CURRENT_DIR)
GET_CSV_FILE = os.path.join(PARENT_DIR, "03.Pandas/kc_house_data _rev.csv")

# If the function you are using doesn't plot lines, there is no need to sort the data:
df_houses = pd.read_csv(GET_CSV_FILE)
```

### **Scatter Plots:**

```python
# Creating a single scatter plot:

sq_lot = df_houses["sqft_lot15"]
prices_area = df_houses["price"]

plt.scatter(sq_lot, prices_area)

plt.xlabel("House Area")
plt.ylabel("Price")

plt.show()
```

```python
# Generate two random datasets on the same plot:

np.random.seed(42)  # For reproducibility
x1 = np.random.randint(low=1, high=11, size=20)
y1 = np.random.randint(20, 150, 20)

x2 = np.random.randint(1, 18, 20)
y2 = np.random.randint(30, 160, 20)

# Create the scatter plot
plt.scatter(x1, y1, color='blue', marker='o', label="X1 & Y1")
plt.scatter(x2, y2, color='red', marker='s', label="X2 & Y2")

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y")
plt.legend()

# Show the plot
plt.show()
```

### **Bar Plots:**

```python
# Group the data by condition and get the sum of all columns:
condition = df_houses.groupby("condition").sum()
condition
```

```python
# Vertical bars
colors = ["yellow", "blue", "green", "orange", "red"]
plt.bar(condition.index, condition["price"], color=colors)
plt.show()
```

```python
# Horizontal bars
plt.barh(condition.index,condition["price"])
plt.show()
```

```python
condition_reset = condition.reset_index()
```

```python
# We will use the replace() method in pandas to replace numerical values with string names.
condition_reset["condition_name"] = condition_reset["condition"].replace({1: "Bad", 2: "Below Average", 3: "Average", 4: "Good", 5: "Excellent"})

condition_reset
```

```python
# Making your bar clear and understandable:
colors = ["yellow", "blue", "green", "orange", "red"]
plt.bar(condition_reset["condition_name"], condition_reset["price"], color=colors)
plt.xlabel("Condition")
plt.ylabel("Price Sum")
plt.show()
```

#### **Multiple Data Bar Plot:**
***1. Grouped Bar Plot (side-by-side comparison):***
<br>Useful when comparing categories across multiple datasets.

```python
labels = ['A', 'B', 'C', 'D']
data1 = [20, 34, 30, 35]
data2 = [25, 32, 34, 20]

x = np.arange(len(labels))  # label locations
width = 0.35                # bar width

fig, ax = plt.subplots()

# First dataset shifted left
bars1 = ax.bar(x - width/2, data1, width, label='Dataset 1')

# Second dataset shifted right
bars2 = ax.bar(x + width/2, data2, width, label='Dataset 2')

ax.set_ylabel('Values')
ax.set_title('Grouped Bar Plot')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

***2. Stacked Bar Plot (cumulative contribution):***
<br>It is useful for showing how datasets contribute to a total.

```python
labels = ['A', 'B', 'C', 'D']
data1 = np.array([20, 34, 30, 35])
data2 = np.array([25, 32, 34, 20])
data3 = np.array([22, 40, 38, 21])

fig, ax = plt.subplots()
ax.bar(labels, data1, label='Dataset 1')
ax.bar(labels, data2, bottom=data1, label='Dataset 2')
ax.bar(labels, data3, bottom=data1 + data2, label='Dataset 3')

ax.set_ylabel('Values')
ax.set_title('Stacked Bar Plot')
ax.legend()

plt.show()
```

### **Histogram Plots:**
- A histogram is a visual representation of frequency distribution.
- It looks similar to a bar chart, but unlike bar charts (which are for categorical data), histograms are specifically for continuous numerical data.
- The x-axis represents intervals (bins) of values, and the y-axis shows the frequency (count) of data points within each bin.

#### **Key Features:**
- Bins (Intervals): Data is divided into consecutive, non-overlapping ranges.
- Bars: Unlike bar charts, histogram bars are adjacent with no gaps, reflecting continuity.
- Shape Matters: The overall shape (normal, skewed, uniform, bimodal) reveals important insights about the data distribution.
- Density Estimation: Histograms can approximate the probability density function of a dataset.

```python
fig, ax = plt.subplots()

ax.hist(df_houses["price"], label="Prices")

ax.set_xlabel("Prices")
ax.set_ylabel("Number of observations")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df_houses["price"], bins=20)

ax.set_xlabel("Prices")
ax.set_ylabel("Number of observations")

plt.show()
```

```python
# Using the "subplots" function to plot multiple histograms on the same chart:
np.random.seed(42)

# Normal distribution data
data_normal = np.random.normal(loc=70, scale=10, size=500)

# Skewed distribution data (exponential)
data_skewed = np.random.exponential(scale=10, size=500) + 50

# Plot histograms
# plt.hist(data_normal, bins=20, label='Normal Distribution', color='blue')
# plt.hist(data_skewed, bins=20, label='Skewed Distribution', color='red')

# The "alpha" parameter in Matplotlib controls the transparency (opacity) of 
# graphical elements such as histograms, scatter plots, or lines.
plt.hist(data_normal, bins=20, alpha=0.6, label='Normal Distribution', color='blue')
plt.hist(data_skewed, bins=20, alpha=0.6, label='Skewed Distribution', color='red')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Normal vs Skewed Distribution')
plt.legend()
plt.show()
```
