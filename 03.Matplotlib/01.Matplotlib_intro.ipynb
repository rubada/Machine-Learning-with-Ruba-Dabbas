#### **Draw our first plot:**

##### **"plt.plot()" function:**
It is used for creating 2D line plots, and drawing lines and points (markers) with various formatting options.
<br>It takes parameters for specifying points in the diagram:
<br>The first parameter is the points on the x-axis.
<br>The second parameter is the points on the y-axis.
<br>The "plt.show()" is a function that displays all figures. It's typically used at the end of your plotting code to render the plots on the screen.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(20)
y = np.array(2*x)

plt.plot(x, y)
plt.show()
```

The keyword argument "marker" is used to emphasize each point with a specified marker.

Check the matplotlib markers link for all possible markers that can be used:
<br>https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

Check the matplotlib specifying colors link for all possible color formats that can be used:
<br>https://matplotlib.org/stable/users/explain/colors/colors.html#specifying-colors

```python
x = np.arange(20)
y = np.sin(x)

# plt.plot(x, y, marker="o")
# plt.plot(x, y, marker="s")
plt.plot(x, y, marker="v")
plt.show()
```

you can change the marker and line colors, linestyle, linewidth, and markersize:

```python
x = np.arange(20)
y = np.sin(x)

plt.plot(x, y, marker="o", color="red", linestyle="--", linewidth=2, markersize=12)
plt.show()
```

The optional parameter "fmt" is a convenient way to define basic formatting, such as color, marker, and linestyle.

```python
x = np.arange(20)
y = np.sin(x)

# plt.plot(x, y, "go:", linewidth=3, markersize=12)
plt.plot(x, y, "go:", lw=3, ms=12)
plt.show()
```

You can use the "mfc" and "mec" to change the marker face color (mfc) and marker edge color (mec).

```python
x = np.arange(20)
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle="-.", marker='o', mfc='red', mec='green')
plt.show()
```

#### **Multiple Lines:**
You can plot multiple lines by simply adding more than one plt.plot() functions:

```python
x = np.arange(6)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='blue', linestyle="-.", marker='o', mfc='red', mec='green')
plt.plot(x, y2, color='c', linestyle="-", marker='o', mfc='m', mec='m')
plt.show()
```

```python
x1 = np.arange(6)
y1 = np.sin(x)

x2 = np.arange(6,12)
y2 = np.cos(x)

plt.plot(x1, y1, x2, y2)
plt.show()
```
